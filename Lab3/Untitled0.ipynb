```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('experiment5.csv', sep=';')
data = data.transpose()
data
```

```python
vavg=[]
x=[]
for i in range(2020):
  vavg.append(np.average(data[i][2:1000]))
  x.append(i%101)
x=x[1:]
vavg=vavg[1:]
```

```python
for size in range(1, 21):
    plt.plot(x[(size-1)*101:(size)*101-1], vavg[(size-1)*101:(size)*101-1])
    plt.show()
```
